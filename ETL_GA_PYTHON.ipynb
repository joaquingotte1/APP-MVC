```python
import pandas as pd
```

```python
from google.analytics.data_v1beta import BetaAnalyticsDataClient
from google.analytics.data_v1beta.types import (
    DateRange,
    Dimension,
    Metric,
    RunReportRequest,
)

def run_report_to_dataframe(property_id="*******", credentials_json_path='C:\\Users\\jgotte\\Desktop\\Anaconda Proyecto\\lithe-key-402312-f56e2eb9aff2.json'):
    client = BetaAnalyticsDataClient.from_service_account_json(credentials_json_path)

    request = RunReportRequest(
        property=f"properties/{property_id}",
        dimensions=[Dimension(name="city"),Dimension(name="country"),Dimension(name="date"), Dimension(name="continent")],
        metrics=[Metric(name="activeUsers"), Metric(name="sessions") , Metric(name="newUsers") , Metric(name="bounceRate") , Metric(name="screenPageViewsPerSession") ],
        date_ranges=[DateRange(start_date="2023-01-01", end_date="2023-11-22")],
    )
    response = client.run_report(request)

    dimensions_city = [row.dimension_values[0].value for row in response.rows]
    dimensions_country = [row.dimension_values[1].value for row in response.rows]
    dimensions_date = [row.dimension_values[2].value for row in response.rows]
    dimensions_continent = [row.dimension_values[3].value for row in response.rows]
    metrics_sesiones = [row.metric_values[1].value for row in response.rows]
    metrics = [row.metric_values[0].value for row in response.rows]
    metrics_NEW_USER = [row.metric_values[2].value for row in response.rows]
    metrics_bounce_rate = [row.metric_values[3].value for row in response.rows]
    metrics_pageviewspersession = [row.metric_values[4].value for row in response.rows]


    data = {'City': dimensions_city, 'country' :dimensions_country, 'date' :dimensions_date,'continent' : dimensions_continent     ,  'sessions' :metrics_sesiones ,
            'ActiveUsers': metrics, 'newUsers': metrics_NEW_USER , 'bounceRate':  metrics_bounce_rate , 'screenPageViewsPerSession':  metrics_pageviewspersession}

   
   
    df = pd.DataFrame(data)

    return df

if __name__ == "__main__":
    result_df = run_report_to_dataframe()
    print("DataFrame result:")
    print(result_df)
```
